### Exemplo 6.7

```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

f = lambda x,y: x-y+2
x0 = 0.0
y0 = 2.0
N = 5

a = 0.0
b = 1.0
X = np.linspace(a,b,N+1)
print (X)
Y = [y0]

h = (b-a)/N

for n in range(len(X)-1):
    k1 = f(X[n], Y[n])
    k2 = f(X[n]+h, Y[n]+h*k1)
    Y.append(Y[n]+0.5*h*(k1+k2))
print (Y)  

s = lambda x: np.exp(-x) + x + 1
Xe = np.linspace(a,b, 20)
Ye = s(Xe)
print (s(X))

# plota
plt.plot(X, Y, "r-", Xe, Ye, "b-")
plt.grid()
plt.show()
```

### Exemplo 6.9

```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

f = lambda x,y: x-y+2
x0 = 0.0
y0 = 2.0
N = 5

a = 0.0
b = 1.0
X = np.linspace(a,b,N+1)
print (X)
Y = [y0]

h = (b-a)/N
for n in range(len(X)-1):
    k1 = f(X[n], Y[n])
    k2 = f(X[n]+0.5*h, Y[n]+0.5*h*k1)
    k3 = f(X[n]+0.75*h, Y[n]+0.75*h*k2)
    Y.append(Y[n] + (h/9) * (2*k1 + 3*k2 + 4*k3))
    #print (k1,Y[-1])
print (Y)

# plota
plt.plot(X, Y, "r-", Xe, Ye, "b-")
plt.grid()
plt.show()
```

```python
% matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

f = lambda x,y: x-y+2
x0 = 0.0
y0 = 2.0
N = 5

a = 0.0
b = 1.0
X = np.linspace(a,b,N+1)
print (X)
Y = [y0]

h = (b-a)/N
for n in range(len(X)-1):
    k1 = f(X[n], Y[n])
    k2 = f(X[n]+0.5*h, Y[n]+0.5*h*k1)
    k3 = f(X[n]+0.5*h, Y[n]+0.5*h*k2)
    
    k4 = f(X[n]+h, Y[n]+h*k3)
    Y.append(Y[n] + (h/6) * (k1 + 2*k2 + 2*k3 + k4))
    #print (k1,Y[-1])
print (Y)

# plota
plt.plot(X, Y, "r-", Xe, Ye, "b-")
plt.grid()
plt.show()
```
